Simple plot using matplotlib 

```python
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns 
```

```python
month=["jan","feb","march","april","may","june"]
y=[5000,4500,8000,7000,9000,3000]
```

```python
plt.plot(figuresize=(5,7))
plt.plot(month,y,color="green",marker="o")
plt.xlabel("month of 2024")
plt.ylabel("expenses")
plt.grid()
plt.title("Every month expenditure of 2024")

plt.show()
```
